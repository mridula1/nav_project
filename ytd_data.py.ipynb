```python
#YTD Formula
'''Calculating the YTD return on a portfolio is the same as for a single investment.
    Take the current value of all assets in the portfolio and subtract the total amount 
    invested on the preceding Jan. 1. This renders the total YTD return in dollars. 
    Dividing this figure by the original value and multiplying by 100 converts it into a
    percentage that reflects the return generated by each dollar originally invested.
'''
# ( Current Price ( end of quarter) - original at beg of year)/(orig at beg of year) * 100
q = month_end.copy()
q.reset_index(level=0, inplace=True)
a = month_beg.copy()
a.reset_index(level=0, inplace=True)

q['Year'] = pd.to_datetime(q.Date).dt.date
a['Year'] = pd.to_datetime(a.Date).dt.date
# a['Date'] = pd.to_datetime(a.Date).dt.month
merged = pd.merge(q,a,left_on ="Year",right_on = "Year", how= 'left')
# merged = pd.merge(q,a,left_on = "Month", right_on = "Month",how ='left')
# # merged = pd.merge(q,a, on= 'Year')
# merged = pd.merge(q,a, on = "Month", how = "left"
# merged = q.merge(a,right_index = True, left_index = True)
merged['Month'] = pd.PeriodIndex(pd.to_datetime(merged.Date_x), freq='M')
# merged['Month'] = pd.PeriodIndex(pd.to_datetime(merged.Date_x), freq='M')
# merged["Month"] = merged["Month"].astype(str)

# Diff is the Closing price month end - closing price month beg /(closing price month beg)
merged["Diff"] = (((merged["Close_x"]-merged["Close_y"]))/(merged["Close_y"])) 

# Percentage change in value from previous month
merged ["NAV_pct_change"] = (merged.Diff.pct_change()) 

df.reset_index(level=0, inplace=True)
df["Month"] = pd.PeriodIndex(pd.to_datetime(df.Date), freq='M')

# df
# month_beg.reset_index(level=0, inplace=True)
# month_end.reset_index(level=0, inplace=True)
# # month_beg['Year'] = pd.to_datetime(a.Date).dt.year.astype(int)
# month_beg["Month"]= pd.PeriodIndex(pd.to_datetime(month_beg.Date), freq='M')
# month_end["Month"]= pd.PeriodIndex(pd.to_datetime(month_end.Date), freq='M')

# month_beg.set_index('Month',inplace = True)
# del month_beg["NAV_change"]
# del month_end["NAV_change"]

# merged["NAV"] = merged["NAV"].astype(float)
# # merged["Close"] = month["Close"]

# # merged["YTD"] = ((merged["Close_x"]-merged["Close_y"])/(merged["Close_y"])) *100
del merged["Open_x"]
del merged["High_x"]
del merged["Low_x"]
del merged["Open_y"]
del merged["High_y"]
del merged["Low_y"]
del merged["Volume_y"]
del merged["Volume_x"]
# del merged["Date_x"]
del merged["Date_y"]
# del merged["Month"]
# del merged["Date"]
del merged["Adj Close_y"]
del merged["Adj Close_x"]

print type(merged["Date_x"][0])
print type(merged["Month"][0])
merged.set_index('Date_x',inplace=True)
merged.index
# # # delete the close 
# merged.to_csv("FFNOX_data.csv")
# month_end.to_csv("FFNOX_Month_End.csv")
# month_beg.to_csv("FFNOX_Month_Beg.csv")
# df.to_csv("FFNOX_df.csv")
# month_end.head
print merged.head
```

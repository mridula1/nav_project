```python
# Code derived from https://www.analyticsvidhya.com/blog/2016/02/time-series-forecasting-codes-python/
def get_stationarity(timeseries_data):
    print timeseries_data.head()
    rolmean = pd.rolling_mean(timeseries_data,window=30)
    print rolmean
    rolstd = pd.rolling_std(timeseries_data,window=12)
    plt.figure()
    timeseries_data.reset_index(level=0, inplace=True)
    orig = plt.plot(timeseries_data["Date"],timeseries_data["Close"],color='blue',label='Original')
    mean = plt.plot(timeseries_data["Date"],rolmean, color='red', label='Rolling Mean')
    std = plt.plot(timeseries_data["Date"],rolstd, color='black', label = 'Rolling Std')
    timeseries_data.set_index('Date',inplace=True)

    plt.legend(loc='best')
    plt.title('Rolling Mean & Standard Deviation')
    plt.savefig("stationarity_fig.png")
```

```python
print df.index
# Eliminating Trend
ts_log = np.log(df)
# line = df.plot(y="Close",use_index=True)
# plt.show()
```

```python
# Process of Decomposing where trend and seasonality are modelled separately and the remaining is returned
print type(ts_log)
print len(ts_log)
print ts_log.index
decomposition = seasonal_decompose(ts_log,freq = 365)

trend = decomposition.trend
seasonal = decomposition.seasonal
residual = decomposition.resid
ts_log_decompose = residual
ts_log_decompose.dropna(inplace=True)
get_stationarity(ts_log_decompose)
```

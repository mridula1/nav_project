```python
new_diff_cols = new.columns
new_diff_cols = list(new_diff_cols)
pct_chng_NAV = pd.DataFrame(index = new.index)
for col in new_diff_cols:
    col_name = col[:6] + "pct_change"
    pct_chng_NAV[col_name] = new[col].pct_change()
```
